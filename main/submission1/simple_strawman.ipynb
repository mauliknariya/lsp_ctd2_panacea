```python
import pandas as pd
import numpy as np
from synapseclient import Synapse
from correlation import gdsc_ccle
user, psswd = open('/Users/mauliknariya/synapse_login.txt')\
                .read().splitlines()
syn = Synapse()
syn.login(user, psswd)
```

#### Differential expression using limma `AveExpr>4.0`

```python
dflimma = pd.read_csv(syn.get('syn22216361').path, compression='gzip')
dflimma.columns = [col.replace('.', '_') for col in dflimma.columns]
dflimma = dflimma[dflimma.AveExpr>4.0]
dflimma
```

#### p-values from limma

```python
genes = sorted(dflimma.hgnc_symbol.unique())
compounds = sorted(dflimma.compound.unique())
dfpval_limma = pd.DataFrame(index = genes)
for compound in compounds:
    pval = pd.Series(dflimma[dflimma.compound==compound]['P_Value'])
    pval.index = dflimma[dflimma.compound==compound]['hgnc_symbol']
    pval.name = compound
    dfpval_limma = pd.concat([dfpval_limma, pval], axis=1, sort=False)
dfpval_limma
```

####  Drugs from DREAM challenge 2019

```python
dfpert19 = pd.read_csv(syn.get('syn22051003').path)
drugs19 = dfpert19.compound.unique().tolist()
drugs19 = [drug for drug in drugs19 
         if drug!='DMSO' and drug!='UNTREATED']
drugs19 = [drug.lower().capitalize() if str.isalpha(drug) is True else drug 
         for drug in drugs19]
```

#### p-values of correlation between IC50 (GDSC) and baseline RNA-seq (CCLE)

```python
from correlation import gdsc_ccle
dfpval_corr = pd.DataFrame()
for drug in drugs19:
    dfcorr = gdsc_ccle(drug)
    pval = pd.Series(dfcorr['pval'])
    if pd.isnull(pval).all():
        continue
    pval.name = drug
    dfpval_corr = pd.concat([dfpval_corr, pval], sort=False, axis=1)
dfpval_corr
```

#### Harmonic mean of p-values (from correlation analysis) across all drugs from 2019

```python
from scipy.stats import hmean, gmean
harmonic = pd.Series(hmean(dfpval_corr.dropna(),axis=1))
harmonic.index = dfpval_corr.dropna().index
harmonic = harmonic[harmonic.index.isin(genes)]
dfpval_limma = dfpval_limma[dfpval_limma.index.isin(harmonic.index)]
```

#### Geometric mean of p-values from limma and p-values from correlation analysis (harmonic mean)

```python
dfgeom = pd.DataFrame(index=harmonic.index)
for compound in compounds:
    df = pd.concat([harmonic, dfpval_limma[compound]],axis=1, sort=False)
    geometric = pd.Series(gmean(df, axis=1))
    geometric.index = harmonic.index
    geometric.name = compound
    dfgeom = pd.concat([dfgeom, geometric], axis=1, sort=False)
dfgeom
```

#### CCLE baseline RNA-seq

```python
dfccle = pd.read_csv(open(syn.get('syn21822697').path), index_col=0)
cells = sorted(dfccle.columns)
num_cells = dfccle.shape[1]
```

#### Ranks based on expression levels of genes with lowest p-values

```python
num_genes = 100
dfpred = pd.DataFrame(index=cells)
for compound in compounds:
    top_genes = dfgeom[compound].sort_values().head(num_genes).index
    dfranks = pd.DataFrame(index=cells)
    for gene in top_genes:
        expr = dfccle.loc[gene, :].sort_values(ascending=False)
        ranks = pd.Series(np.arange(1, num_cells+1, dtype=int))
        ranks.index = expr.index
        ranks.name = gene
        dfranks = pd.concat([dfranks, ranks], sort=False, axis=1)
    dfpred[compound] = dfranks.median(axis=1)
dfpred = dfpred.apply(lambda x: (dfccle.shape[1]-x)/dfccle.shape[1])
dfpred
```

```python
dfgdsc = pd.read_csv(syn.get('syn22051024').path)
dfgdsc['IC50'] = dfgdsc['LN_IC50'].apply(lambda x: np.exp(x))
dftrue = pd.DataFrame(index=cells)
for drug in drugs19:
    dfic50 = dfgdsc[(dfgdsc.DRUG_NAME==drug) &
                    (dfgdsc.CELL_LINE_NAME.isin(cells))]\
                   [['CELL_LINE_NAME', 'IC50']]
    if dfic50.empty is True: continue
    dfic50.index = dfic50['CELL_LINE_NAME']
    dfic50 = dfic50.drop(columns='CELL_LINE_NAME')
    dfic50 = dfic50.groupby(dfic50.index).mean()
    dfic50 = dfic50.sort_values(by='IC50', ascending=False)
    ranks = pd.Series(np.arange(1, dfic50.shape[0]+1, dtype=int),
                     index=dfic50.index, name=drug)
    dftrue = pd.concat([dftrue, ranks], sort=False, axis=1)
dftrue = dftrue.apply(lambda x: (num_cells-x)/num_cells)
dftrue.columns = [col.upper() for col in dftrue.columns]
dftrue
```

```python
import seaborn as sns
sns.set(style='darkgrid', font_scale=1.25)
import matplotlib.pyplot as plt

dfplot = pd.DataFrame()
for drug in dftrue.columns:
    dfplot = dfplot.append(pd.DataFrame
                           (list(zip([drug]*num_cells,
                                     ['measured']*num_cells,
                                     dftrue[drug], dfpred[drug])), 
                            columns=['drug', 'result', 'measured', 'predicted']
                           ))
grd = sns.FacetGrid(dfplot, col='drug', 
                    col_wrap=5,
                    sharex=True, sharey=True)
grd.map(sns.scatterplot, 'measured', 'predicted')
```
